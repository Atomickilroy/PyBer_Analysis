```python
#Import Dependences
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#Create Bubble Plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s =[i*5 for i in y_axis])
#Incorporate Parameters

#Add legend-Boston  
#Color sky blue
#Size 5 times 
#Color 20% transparent
#Add black edging to the circles
#Make linewidth of circles 2 points
#Switch the axis so that the Fare($) data is on the x-axis

ax.scatter(y_axis, x_axis, s =[i*5 for i in y_axis], facecolor="skyblue",
           edgecolor= "black",linewidth=2,label =('Boston'), alpha=0.8)

#Create Legend
ax.legend()

#Create Title
ax.set_title("PyBer Fare by Month")

#Invert the y_axis
plt.gca().invert_yaxis()           

#Create a label for the x and y axes 
ax.set_ylabel("Date")             
ax.set_xlabel("Fare($)")             

#Add a limit on x_axis
ax.set_xlim(0,50)

#Print 
plt.show()
```
