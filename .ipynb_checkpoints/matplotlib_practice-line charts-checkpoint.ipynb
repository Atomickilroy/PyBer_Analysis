```python
#Import dependencies

import matplotlib.pyplot as plt

#Import the statistics module to the sheet

import statistics
```

```python
#MATHLAB 
```

```python
# Set the x-axis to a list of strings for each month

x_axis = ["Jan", "Feb", "Mar", "April","May","June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"] 
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Calculate the standard deviation of the y_axis value using statistics.stdev()

stdev = statistics.stdev(y_axis)
stdev
```

```python
#For the Line Chart add stdev, x,y axes to errorbar()

plt.errorbar(x_axis, y_axis, yerr = stdev,)
```

```python
#add cap tp the error bar that's equal to a scalar value using capsize= parameter

plt.errorbar(x_axis,
             y_axis,
             yerr = stdev,
             capsize=3,
            
            )
```

```python
#Here, the standard deviations above and below the mean overlap with the error bars
#from some of the other months. When comparing all the months like this, we can assume 
#that there is very little variability in the fares from month to month.
#And overall, there is a wide range in the average fare price for every month.
```

```python
#Object Oriented Interface Method
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis,
           y_axis,
           yerr=stdev,
           capsize =3,
           
           )
plt.show()
```

```python
#Adding Errorbars to a Bar Chart
#We can use the same parameters as above
```

```python
#MATLAB Method

plt.bar(x_axis,
           y_axis,
           yerr=stdev,
           capsize =3,
        
       )
```

```python
#Object Oriented Interface Method
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis,
           y_axis,
           yerr=stdev,
           capsize =3,
      
      )
plt.show()
```

```python
#Change the Major Ticks 
#To adjust for the major ticks use xticks() but first  import numpy

import numpy as np
```

```python
#Now rechart the barh and include the xticks() and space out to $5 increments

plt.barh(x_axis, y_axis, )
plt.xticks(np.arange(0, 51, step=5.0))

#Invert y-axis
plt.gca().invert_yaxis()
```

```python
#What is step = doing?

import numpy as np

np.arange (0,51, step =5.0)
```

```python
#Object Oriented Interface Method
```

```python
#Create the same graph
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)

#Dont forget to include ax.set_ before to the xticks()
ax.set_xticks(np.arange(0,51 , step= 5.0))

plt.show()
```

```python
#Add Minor ticks to the OOI barh chart for the matplotlib.ticker and import MultipleLocator
```

```python
from matplotlib.ticker import MultipleLocator

#Increase the size of the plot figure
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

#Create the minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.show()

```

```python
#Now, the figure will have minor ticks every $1 increment and major ticks at every $5 increment
```

```python
#5.1.4 Annotate the Chart
```

```python
#Create the plot and add a label for the legend
#Additional parameters added to plt.plot() for line style/color

plt.plot(x_axis, y_axis, marker="*", color="orange", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Annotate the plot with the Object Oriented Method

fig = plt.figure()

ax = fig.add_subplot()

#Skill Drill 5.1.4

# Change the line to green, width of 2, marker diamond, and label Boston
ax.plot(x_axis, y_axis, color="green", marker = "d",linewidth = 2 ,label= "Boston")

#Add Legend
ax.legend()

#Set Title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare(s)")
ax.set_ylabel("Date")

#Set y Limit
ax.set_ylim(0,45)

#Add Grid lines
ax.grid()
```
